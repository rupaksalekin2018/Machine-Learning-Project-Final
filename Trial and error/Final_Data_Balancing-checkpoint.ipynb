```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
dir = "C:\\Users\\User\\Desktop\\cataract\\Training"
```

```python
categories = ['normal', 'cataract']
```

```python
data = []
```

```python
for category in categories:
    path = os.path.join(dir,category)
    label = categories.index(category)
    
    for img in os.listdir(path):
        imgpath = os.path.join(path,img)
        dis_img = cv2.imread(imgpath,0)
        dis_img = cv2.resize(dis_img,(200,200))
        image = np.array(dis_img).flatten()
        
        data.append([image,label])
```

```python
import random
```

```python
random.shuffle(data)
features = []
labels = []
```

```python
for feature, label in data:
    features.append(feature)
    labels.append(label)
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(features, labels, test_size=0.20, random_state=4)
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE()
x_train_resampled, y_train_resampled = smote.fit_resample(xtrain,ytrain)
```

```python
x_test_resampled, y_test_resampled = smote.fit_resample(xtest,ytest)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import plot_confusion_matrix
import seaborn as sns
```

```python
clf = LogisticRegression().fit(xtrain,ytrain)
y_predict = clf.predict(xtest)
sns.set_context("poster")
disp = plot_confusion_matrix(clf, xtest, ytest, cmap='cividis', colorbar=False)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(ytest, y_predict))
```

```python
pred_clf = clf.predict(xtest)
misclassified=np.where(ytest!=pred_clf)
print("Total Misclassified Samples: ",len(misclassified[0]))
```
