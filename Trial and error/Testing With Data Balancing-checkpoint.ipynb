```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
dir = "C:\\Users\\User\\Desktop\\cataract\\Training"
```

```python
categories = ['normal', 'cataract']
```

```python
normal_data = []
cataract_data = []
```

```python
for category in categories:
    path = os.path.join(dir,category)
    label = categories.index(category)
    if label == 0:
        for img in os.listdir(path):
            imgpath = os.path.join(path,img)
            dis_img = cv2.imread(imgpath,0)
            dis_img = cv2.resize(dis_img,(200,200))
            image = np.array(dis_img).flatten()
            normal_data.append([image,label])
    if label == 1:
        for img in os.listdir(path):
            imgpath = os.path.join(path,img)
            dis_img = cv2.imread(imgpath,0)
            dis_img = cv2.resize(dis_img,(200,200))
            image = np.array(dis_img).flatten()
            cataract_data.append([image,label])
```
