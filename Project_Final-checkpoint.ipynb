```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
dir = "C:\\Users\\User\\Desktop\\FINAL_DATASET"
```

```python
categories = ['normal', 'cataract']
```

```python
data = []
```

```python
for category in categories:
    path = os.path.join(dir,category)
    label = categories.index(category)
    
    for img in os.listdir(path):
        imgpath = os.path.join(path,img)
        dis_img = cv2.imread(imgpath,0)
        dis_img = cv2.resize(dis_img,(200,200))
        image = np.array(dis_img).flatten()
        
        data.append([image,label])
```

```python
import random
```

```python
random.shuffle(data)
features = []
labels = []
```

```python
for feature, label in data:
    features.append(feature)
    labels.append(label)
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(features, labels, test_size=0.20, random_state=4)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import plot_confusion_matrix
import seaborn as sns
```

```python
clf = LogisticRegression().fit(xtrain,ytrain)
y_predict = clf.predict(xtest)
sns.set_context("poster")
disp = plot_confusion_matrix(clf, xtest, ytest, cmap='cividis', colorbar=False)
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest, y_predict))
```

```python
from sklearn.svm import SVC
svc_model = SVC(C=1, probability=True, kernel="poly", gamma="auto").fit(xtrain,ytrain)
y_svc_predict = svc_model.predict(xtest)
sns.set_context("poster")
disp = plot_confusion_matrix(svc_model, xtest, ytest, cmap='cividis', colorbar=False)
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest, y_svc_predict))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
random_forest = RandomForestClassifier(n_estimators=100).fit(xtrain,ytrain)
y_rf_predict = random_forest.predict(xtest)
sns.set_context("poster")
disp = plot_confusion_matrix(random_forest, xtest, ytest, cmap='cividis', colorbar=False)
```

```python
print(classification_report(ytest, y_rf_predict))
```

```python
from sklearn.ensemble import AdaBoostClassifier
abc = AdaBoostClassifier(n_estimators=50,learning_rate=1).fit(xtrain,ytrain)
y_abc_predict = abc.predict(xtest)
sns.set_context("poster")
disp = plot_confusion_matrix(abc, xtest, ytest, cmap='cividis', colorbar=False)
```

```python
print(classification_report(ytest, y_abc_predict))
```

```python
from sklearn.ensemble import StackingClassifier, RandomForestClassifier, AdaBoostClassifier
from sklearn.linear_model import LogisticRegression 
from sklearn.svm import SVC
base_learners = [
                 ('rf_1', RandomForestClassifier(n_estimators=10, random_state=42)),
                 ('rf_2', SVC(C=1, probability=True, kernel="poly", gamma="auto")),
                ('rf_3', AdaBoostClassifier(n_estimators=50, learning_rate=1))
                ]
clf_final_stack = StackingClassifier(estimators=base_learners, final_estimator=LogisticRegression())
clf_final_stack.fit(xtrain, ytrain)
y_stack_predict = clf_final_stack.predict(xtest)
sns.set_context("poster")
```

```python
disp = plot_confusion_matrix(clf_final_stack, xtest, ytest, cmap='cividis', colorbar=False)
```

```python
print(classification_report(ytest, y_stack_predict))
# clf_final_stack.score(xtest,ytest)
```

```python
r_probs = [0 for _ in range(len(ytest))]
clf_probs = clf.predict_proba(xtest)
svc_probs = svc_model.predict_proba(xtest)
rf_probs = random_forest.predict_proba(xtest)
abc_probs = abc.predict_proba(xtest)
stack_probs = clf_final_stack.predict_proba(xtest)
```

```python
clf_probs = clf_probs[:, 1]
svc_probs = svc_probs[:, 1]
rf_probs = rf_probs[:, 1]
abc_probs = abc_probs[:, 1]
stack_probs = stack_probs[:, 1]
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
r_auc = roc_auc_score(ytest, r_probs)
clf_auc = roc_auc_score(ytest, clf_probs)
svc_auc = roc_auc_score(ytest, svc_probs)
rf_auc = roc_auc_score(ytest, rf_probs)
abc_auc = roc_auc_score(ytest, abc_probs)
stack_auc = roc_auc_score(ytest, stack_probs)
```

```python
print('Random (chance) Prediction: AUROC = %.3f' % (r_auc))
print('Random Forest: AUROC = %.3f' % (rf_auc))
print('Logistic Regression: AUROC = %.3f' % (clf_auc))
print('SVM : AUROC = %.3f' % (svc_auc))
print('Ada Boost : AUROC = %.3f' % (abc_auc))
print('Stack: AUROC = %.3f' % (stack_auc))
```

```python
r_fpr, r_tpr, _ = roc_curve(ytest, r_probs)
rf_fpr, rf_tpr, _ = roc_curve(ytest, rf_probs)
clf_fpr, clf_tpr, _ = roc_curve(ytest, clf_probs)
abc_fpr, abc_tpr, _ = roc_curve(ytest, abc_probs)
svc_fpr, svc_tpr, _ = roc_curve(ytest, svc_probs)
stack_fpr, stack_tpr, _ = roc_curve(ytest, stack_probs)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(r_fpr, r_tpr, linestyle='--', label='Random prediction (AUROC = %0.3f)' % r_auc)
plt.plot(rf_fpr, rf_tpr, marker='.', label='Random Forest (AUROC = %0.3f)' % rf_auc)
plt.plot(abc_fpr, abc_tpr, marker='.', label='Ada Boosting (AUROC = %0.3f)' % abc_auc)
plt.plot(svc_fpr, svc_tpr, linestyle='--', label='SVM (AUROC = %0.3f)' % svc_auc)
plt.plot(stack_fpr, stack_tpr, marker='.', label='Stack (AUROC = %0.3f)' % stack_auc)
plt.plot(clf_fpr, clf_tpr, marker='.', label='Logistic Regression (AUROC = %0.3f)' % clf_auc)

# Title
plt.title('ROC Plot')
# Axis labels
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
# Show legend
plt.legend() # 
# Show plot
plt.rcParams["figure.figsize"] = (20,5)
plt.show()
```
